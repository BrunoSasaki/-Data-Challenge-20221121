```python
import pandas as pd
```

```python
dfn = pd.read_csv('/content/netflix_titles.csv')
display(dfn)
```

```python
dfa = pd.read_csv('/content/amazon_prime_titles.csv')
display(dfa)
```

```python
df = pd.concat([dfn, dfa])

# contando o número de vezes que cada país aparece
country_counts = df.groupby("country").size()

# obtendo os 5 países com mais ocorrências
top_countries = country_counts.sort_values(ascending=False).head(5)

# imprimindo os resultados
print("Top 5 países produtores de conteúdo:")
print(top_countries)
```

```python
actor_counts = df["cast"].value_counts()

# obtendo os 10 atores/atrizes com mais ocorrências
top_actors = actor_counts.head(10)

# imprimindo os resultados
##print("Top 10 atores/atrizes:")
print(top_actors)
```

```python
dfa["date_added"] = pd.to_datetime(dfa["date_added"])
dfa.info()
```

```python
dfa["month_added"] = dfa["date_added"].dt.month
dfa
```

```python
# contando o número de vezes que cada mês aparece
month_counts = dfa["month_added"].value_counts()
```

```python
most_common_month = month_counts.index[0]
```

```python
print("Mês com mais adições de filmes na Netflix:", most_common_month)
```

```python
# contando a quantidade de filmes com gênero "Comédia"
comedy_count = df[df["listed_in"].str.contains("Comedy", na=False)].shape[0]

# imprimindo o resultado
print("Quantidade de filmes listados como comédia:", comedy_count)
```

```python

# criando uma lista com todos os gêneros de filmes
genres = set()
for genres_string in df["listed_in"]:
    genres_list = genres_string.split(",")
    for genre in genres_list:
        genres.add(genre.strip())

# imprimindo a lista de gêneros na tela
print("Lista de todos os gêneros de filmes:")
for genre in sorted(genres):
    print(genre)
```
