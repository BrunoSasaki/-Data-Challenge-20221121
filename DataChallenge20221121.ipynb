```python
# Resolução do Data Challenge para a Vaga de Data Analyst

# Importação da biblioteca pandas para manipulação de dados

import pandas as pd

```

```python
#Realizando a Leitura do arquivo csv da Netflix e salvando em uma variável chamada dfn 

dfn = pd.read_csv('/content/netflix_titles.csv')
display(dfn)
```

```python
#Realizando a Leitura do arquivo csv da Netflix e salvando em uma variável chamada dfa

dfa = pd.read_csv('/content/amazon_prime_titles.csv')
display(dfa)
```

```python
#Concatenando as duas tabelas e uma variável chamada df

df = pd.concat([dfn, dfa])
```

```python
# 1- Top 10 atores/atrizes considerando todos os dados;

#Contando o número de vezes que cada ator/atriz aparece na coluna "cast" com o método value_counts() e salvando em uma variável chamada actor_counts

actor_counts = df["cast"].value_counts()

# obtendo os 10 atores/atrizes com mais ocorrências
top_actors = actor_counts.head(10)

# imprimindo os resultados
print("Top 10 atores/atrizes:")
print(top_actors)
```

```python
# 2- Top 5 países produtores de conteúdos considerando todos os dados e comparando as duas plataformas;

# agrupando os dados por País utilizando a coluna "country" através do método groupby()
country_counts = df.groupby("country").size()

# obtendo os 5 países com mais ocorrências através do método sort_values() e colocando em ordem decrescente
top_countries = country_counts.sort_values(ascending=False).head(5)

# imprimindo os resultados
print("Top 5 países produtores de conteúdo:")
print(top_countries)
```

```python
# 3- Mês no qual há mais adições de filmes na plataforma Netflix;

#Utilizando o método to_datetime do pandas podemos alterar o dado na coluna "date_added" que está como object para datetime e salvando no própria variável
#Nesse caso como queremos os dados somente da netflix utilizaremos a variável dfn
dfn["date_added"] = pd.to_datetime(dfn["date_added"])
dfn.info()
```

```python
#Extraindo o mês da coluna "date_added" já transformada e criando uma nova coluna chamada "month_added"
dfn["month_added"] = dfn["date_added"].dt.month
dfn
```

```python
# contando o número de vezes que cada mês aparece
month_counts = dfn["month_added"].value_counts()

#retorna uma série com todos os índices da série resultante, ordenados pelo número de ocorrências em ordem decrescente. Portanto, index[0]
#significa o primeiro índice da série resultante, ou seja, o índice com o maior número de ocorrências na série original

most_common_month = month_counts.index[0]

# imprimindo o resultado

print("Mês com mais adições de filmes na Netflix:", most_common_month)
```

```python
# 4- Quantidade de filmes listados como comédia;

# filtrando as linhas que contém o genêro "Comedy" dentro da coluna "listed_in", utilizando o método str.contains() e salvando na variável comedy_count
# O argumento na=False foi utilizado para tratar valores faltantes como strings vazias em seguida o método shape.[0] para contar o número de linhas filtradas

comedy_count = df[df["listed_in"].str.contains("Comedy", na=False)].shape[0]

# imprimindo o resultado
print("Quantidade de filmes listados como comédia:", comedy_count)
```

```python
# 5- Lista de todos os gêneros de filmes;

# criando uma lista com todos os gêneros de filmes

# criando um conjunto vazio chamado genres para armazenar os gêneros de filmes únicos. Ele percorre cada string de gêneros na coluna "listed_in" e usa
#o método split() para separar a string em uma lista de gêneros individuais. Em seguida, ele percorre cada gênero na lista e usa o método strip() para 
#remover espaços em branco adicionais. 
#Adicionando cada gênero único ao conjunto genres. Utilizando o método sorted() para ordenar a lista de gêneros em ordem alfabética 

genres = set()
for genres_string in df["listed_in"]:
    genres_list = genres_string.split(",")
    for genre in genres_list:
        genres.add(genre.strip())

# imprimindo a lista de gêneros na tela
print("Lista de todos os gêneros de filmes:")
for genre in sorted(genres):
    print(genre)
```

```python
# 6- A frequência de "TV Show" de todos os dados e comparativamente em relação as duas plataformas;

# contar o número de ocorrências da string "TV Show" na coluna "type". utilizando o método shape[0] para contar o número de linhas que contêm a 
#string "TV Show" na coluna "type" de cada DataFrame. Ele armazena a frequência de "TV Show" na Netflix na variável netflix_tv_shows e a frequência 
#de "TV Show" na Amazon na variável amazon_tv_shows. 

# contando a frequência de "TV Show" na Netflix
netflix_tv_shows = dfn[dfn["type"] == "TV Show"].shape[0]


# imprimindo os resultados na tela
print("Frequência de 'TV Show' na Netflix:", netflix_tv_shows)
```

```python
# contando a frequência de "TV Show" na Amazon
amazon_tv_shows = dfa[dfa["type"] == "TV Show"].shape[0]

# imprimindo os resultados na tela

print("Frequência de 'TV Show' na Amazon:", amazon_tv_shows)
```

```python
#7- A frequência de "Movies" de todos os dados e comparativamente em relação as duas plataformas;

# contando a frequência de "Movie" na Netflix
netflix_movies = dfn[dfn["type"] == "Movie"].shape[0]

# contando a frequência de "Movie" na Amazon
amazon_movies = dfa[dfa["type"] == "Movie"].shape[0]

# imprimindo os resultados na tela
print("Frequência de 'Movie' na Netflix:", netflix_movies)
print("Frequência de 'Movie' na Amazon:", amazon_movies)
```
